### Question 1: GET and POST method:
(i) In GET - parameters are visible in the URL , in POST - parameters are not visible in URL.
(ii) Caching is possible in GET, while caching is not possible in POST.
(iii) The request remains in browser history when we use GET, but not available when we use POST.
(iv) The request can be bookmarked in GET, but cannot be bookmarked in POST.
(V) Not suitable for sensitive data incase of Get, but suitable for sensitive data in POST.
(vi) If we use GET, there is limitation on length of data, but there is no limitation on length of data in POST.
(vii) GET is mostly used for requesting a resource while POST is used for creating a resource.

### Question 5: Flask app with Postman:

```python
from flask import Flask,request

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello!!"

@app.route("/user",methods=['POST'])
def hello_user():
    return f"Hello {request.form.get('name')}!!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
